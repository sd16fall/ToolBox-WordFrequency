```python
import string
def process(bookname):
    book = open(bookname)
    dic = dict()
    for word in book:
        for wordss in word.split():
            wordsss = wordss.strip(string.punctuation+string.whitespace)
            wordsss = wordsss.lower()
        if wordsss in dic:
            dic[wordsss] +=1
        else:
            dic[wordsss] =1
    tl=list()
    for key in dic:
        tl.append((dic[key],key))
    tl.sort(reverse=True)
    tll=tl[0:100]
    for (a,b) in tll:
        print b
    

process("pg848.txt")
```
